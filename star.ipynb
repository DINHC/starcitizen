```python
#import the neccessary dependencies
import pandas as pd
import requests
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
#create a variable for your URl
url = "https://starcitizen.fandom.com/wiki/List_of_ships"

#create a variable for your requests to the url
r = requests.get(url)


#read the url 
df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list

#create a dataframe with the table. In my case it was the only table on the web page
df = df_list[0]
df.head()
```
